```python
data["age"].mean()
```

```python
data["age"].mode()
```

###### here mean is greater than median so it shows that outliers are present in age. so we choose median to fill missing values of age

```python
data.fillna(data.median(), inplace = True)
```

```python
sns.boxplot(data["age"])
```

###### By observing this, there are outliers present in age. we need to identify the rows in age showing outlier values using IQR Method

```python
q1_age = data["age"].describe()["25%"]
q3_age = data["age"].describe()["75%"]
```

```python
q1_age
```

```python
q3_age
```

```python
iqr_age = q3_age - q1_age
```

```python
iqr_age
```

```python
lower_boundary = q1_age - (1.5 * iqr_age)
```

```python
upper_boundary = q3_age + (1.5 * iqr_age)
```

```python
lower_boundary
```

```python
upper_boundary
```

```python
data[data["age"] < lower_boundary]
```

```python
(data[data["age"] < lower_boundary].shape[0] / data["age"].shape[0]) * 100
```

###### This value shows the percentage of outliers in lower boundary which is approximaatly 3%.so we can neglect these values

```python
data[data["age"] > upper_boundary]
```

```python
(data[data["age"] > upper_boundary].shape[0] / data["age"].shape[0]) * 100
```

###### This value shows the percentage of outliers in upper boundary which is approximaatly 5%.so we can neglect these values
